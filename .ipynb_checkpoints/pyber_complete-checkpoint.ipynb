```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data_to_load)
ride = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride = pd.merge(ride, city, on="city")

# Display the data table for preview
city_ride.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_ride["rides"] = 1
urban = city_ride.loc[city_ride["type"] == "Urban"]
urbanbycity = pd.pivot_table(urban, values=["rides","driver_count", "fare"],index='city',aggfunc = sum)
urbanbycity["driver_count"] = urbanbycity["driver_count"] / urbanbycity["rides"]
urbanbycity["avgfare"] = urbanbycity["fare"] / urbanbycity["rides"]

suburban = city_ride.loc[city_ride["type"] == "Suburban"]
suburbanbycity = pd.pivot_table(suburban, values=["rides","driver_count", "fare"],index='city',aggfunc = sum)
suburbanbycity["driver_count"] = suburbanbycity["driver_count"] / suburbanbycity["rides"]
suburbanbycity["avgfare"] = suburbanbycity["fare"] / suburbanbycity["rides"]

rural = city_ride.loc[city_ride["type"] == "Rural"]
ruralbycity = pd.pivot_table(rural, values=["rides","driver_count", "fare"],index='city',aggfunc = sum)
ruralbycity["driver_count"] = ruralbycity["driver_count"] / ruralbycity["rides"]
ruralbycity["avgfare"] = ruralbycity["fare"] / ruralbycity["rides"]

#Build the scatter plots for each city types
c_scale = 10
urban = plt.scatter(x=urbanbycity["rides"],y=urbanbycity["avgfare"],s=urbanbycity["driver_count"]*c_scale,c='coral',alpha=0.75,edgecolors="black")
suburban = plt.scatter(x=suburbanbycity["rides"],y=suburbanbycity["avgfare"],s=suburbanbycity["driver_count"]*c_scale,c='skyblue',alpha=0.75,edgecolors="black")
rural = plt.scatter(x=ruralbycity["rides"],y=ruralbycity["avgfare"],s=ruralbycity["driver_count"]*c_scale,c='gold',alpha=0.75,edgecolors="black")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare ($)")
plt.grid(True)

# Create a legend
leg = plt.legend((urban, suburban, rural),("urban", "suburban", "rural"),loc='best',ncol=1,fontsize=8,markerscale=1 )
leg.legendHandles[0]._sizes=[30]
leg.legendHandles[1]._sizes=[30]
leg.legendHandles[2]._sizes=[30]

# Incorporate a text label regarding circle size
plt.text(5, 12, 'Note: Circle size correlates with driver count per city')

# Save Figure
plt.savefig("Images/avg_fare_bynumrides.png")
plt.show()
urbanbycity.head()
```

## Total Fares by City Type

```python
# Calculate Type Percents
cd_type = city_ride.groupby("type")
a = cd_type["fare"].sum()
b_amt = [a[0], a[1], a[2]]
b_labels = a.keys()

# Create a pie chart based upon the trip duration of that single bike
explode = (0, 0, 0.1)
colors = ["gold", "skyblue", "coral"]

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
ax.pie(a,explode=explode, colors=colors, labels = b_labels, startangle=140, autopct="%1.1f%%", shadow=True)
ax.set_title("% of Total Fares by City Type")

# Save Figure
plt.savefig("Images/fares_bytype.png")
plt.show()
```

```python
# Calculate Type Percents
#cd_type = city_ride.groupby("type")
a = cd_type["type"].count()
b_amt = [a[0], a[1], a[2]]
b_labels = a.keys()

# Create a pie chart based upon the trip duration of that single bike
explode = (0, 0, 0.1)
colors = ["gold", "skyblue", "coral"]

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))
ax.pie(b_amt,explode=explode, labels=b_labels, colors=colors, startangle=140, autopct="%1.1f%%", shadow=True)
ax.set_title("% of Total Rides by City Type")

# Save Figure
plt.savefig("Images/rides_bytype.png")
plt.show()
```
